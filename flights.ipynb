draw the bar chat year vs passengers

```python

plt.figure(figsize=(10, 6))
plt.bar(df['year'].astype(str), df['passengers'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Total Passengers per Year')
```

draw the bar chat for all months in a particular year

```python

colors = ['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightblue',
          'lightyellow', 'lightcyan', 'lavender', 'lightpink', 'lightgray',
          'lightgoldenrodyellow', 'whitesmoke']
specific_year = 1995
plt.figure(figsize=(10, 6))
plt.bar(df['month'], df['passengers'], color=colors)
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.title(f'Passengers per Month in {specific_year}')
```

draw the pie flot year vs passengers

```python
k = df.head(20)
labels = k['year'].astype(str)
sizes = k['passengers']
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Passenger Distribution by Year')
plt.axis('equal')
plt.legend()
plt.show()
```

draw scatter plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['year'], df['passengers'], color='coral', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Scatter Plot of Passengers vs. Year')
```
